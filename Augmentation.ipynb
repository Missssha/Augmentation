```python
def augment_image(image_path, points):
    # Загружаем изображение и JSON
    img = cv2.imread(image_path)
    
    # Выбираем случайный тип аугментации
    augmentation_type = random.choice(['ogging'])
    # augmentation_type = "rotation"            'gorizontal_flip', 'vertical_flip', 'lightening', 'ogging'
    

    # Применяем аугментацию
    #поворот
    if augmentation_type == 'rotation':
        # angle = random.uniform(-30.0, 30.0)
        angle = random.uniform(-15.0, 15.0)
        # Центр вращения
        height, width = img.shape[:2]
        center = (width // 2, height // 2)

        # Матрица поворота
        rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)

        # Поворот изображения
        img = cv2.warpAffine(img, rotation_matrix, (width, height))

        # Обновление координат точек
        rot_points = []
        for point in points:
            x, y = point
            # Преобразование координат с помощью матрицы поворота
            new_x, new_y = np.dot(rotation_matrix[:2, :2], np.array([x, y]) - center) + center
            rot_points.append([int(new_x), int(new_y)])
        points = rot_points[:]

    # elif augmentation_type == 'scaling':
    #     scale_factor = random.uniform(0.5, 2.0)

    #     height, width = img.shape[:2]
    #     new_height = int(height * scale_factor)
    #     new_width = int(width * scale_factor)

    #     # Масштабирование изображения
    #     img = cv2.resize(img, (new_width, new_height))

    #     # Обновление координат точек
    #     scal_points = []
    #     for point in points:
    #         x, y = point
    #         scaled_x = int(x * scale_factor)
    #         scaled_y = int(y * scale_factor)
    #         scal_points.append([scaled_x, scaled_y])
    #     points = scal_points[:]

    #горизонтальное отражение
    elif augmentation_type == 'gorizontal_flip':
        height, width = img.shape[:2]

        # Горизонтальное отражение
        img = cv2.flip(img, 1) 

        # Обновление координат точек
        flip_points = []
        for point in points:
            x, y = point
            # Новая координата x: ширина изображения - исходная координата x
            flipped_x = width - x - 1
            flip_points.append([flipped_x, y])
        points = flip_points[:]
        
    #вертикальное отражение
    elif augmentation_type == 'vertical_flip':
        height, width = img.shape[:2]

        # Вертикальное отражение
        img = cv2.flip(img, 0) 

        # Обновление координат точек
        flip_points = []
        for point in points:
            x, y = point
            # Новая координата y: высота изображения - исходная координата y
            flipped_y = height - y - 1
            flip_points.append([x, flipped_y])
        points = flip_points[:]

    #засветление
    elif augmentation_type == 'lightening':
        gamma = random.uniform(2.0, 2.8)
        gamma = 2.8
        invGamma = 1.0 / gamma
        table = np.array([((i / 255.0) ** invGamma) * 255
            for i in np.arange(0, 256)]).astype("uint8")
        img = cv2.LUT(img, table)

    #затемнение
    elif augmentation_type == 'ogging':
        gamma = random.uniform(0.5, 0.7)
        gamma = 0.3
        invGamma = 1.0 / gamma
        table = np.array([((i / 255.0) ** invGamma) * 255
            for i in np.arange(0, 256)]).astype("uint8")
        img = cv2.LUT(img, table)
    # print(points)

    
    #выбор последнего файла
    # name_of_last_file = os.listdir('C:/Users/PlusninMA/Desktop/accumulator_test/section/section_augmentation')[-1] 
    # number_of_last_file = int(name_of_last_file.split("_")[0]) + 1
    number_of_last_file = len(os.listdir('C:/Users/PlusninMA/Desktop/camera_test/section/section_train')) - 50

    #создание имени нового файла
    # image_file_name = str(number_of_last_file) + "_augmented_" + image_path.split("/")[-1] 
    image_file_name = str(number_of_last_file + 1) + "_augmented_" + image_path.split("/")[-1] 
    image_path_to_save = 'C:/Users/PlusninMA/Desktop/camera_test/section/section_train/' + image_file_name
    
    # Сохраняем измененное изображение и обновляем JSON
    cv2.imwrite(image_path_to_save, img)

    output_image_path = "/".join(image_path.split("/")[:-1]) + image_file_name
    output_json_path = os.path.splitext(output_image_path)[0] + '.json'

    # Заполняем новый JSON файл
    data = {
            "filename": os.path.basename(image_path_to_save),
            "size": os.path.getsize(image_path_to_save),
            "regions": [],
            "file_attributes": {
                "caption" : "",
                "public_domain" : "no",
                "image_url" : ""
            }
        }
    regions = []
    for region in range(len(points) // 4):
        regions.append({
                    "shape_attributes": {
                        "name": "polygon",
                        "all_points_x": [point[0] for point in points[region*4:(region+1)*4]],
                        "all_points_y": [point[1] for point in points[region*4:(region+1)*4]]
                    },
                    "region_attributes": {
                        "name" : "not_defined",
                        "type" : "unknown",
                        "image_quality" : {
                            "good" : "true",
                            "frontal" : "true",
                            "good_illumination" : "true"
                        }
                    }
                })
    
    data["regions"] = regions
    # print(data)
    #обновление данных json файла
    old_data = json.load(open(augmentation_json_path))
    old_data[image_file_name] = data
    # print(old_data)
    with open(augmentation_json_path, "w") as file:
        json.dump(old_data, file, indent=4, ensure_ascii=False)

    return image_file_name

    


```
